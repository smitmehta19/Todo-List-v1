### Task 6 : Prediction using Decision Tree Algorithm

### Problem Statement: We are given some features (sepal length, sepal width, petal length, petal width) of a flower, and based on these features we have to identify the category in which the flower belongs(setosa, virginica or versicolor).

### Dataset : Iris

```
# Importing required libraries

import numpy as np 
import pandas as pd
import graphviz

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score 

from sklearn.datasets import load_iris
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn import tree
```

```
# Loading the iris dataset
iris=datasets.load_iris()
```

```
#printing features name of iris data 

print ("Features Name : ", iris.feature_names)
```

```
# Forming the iris dataframe
df=pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head(5))
```

```
df.describe()
```

```
#shape of datasets 

print ("Dataset Shape: ", df.shape)
```

```
#printing samples and target 

X = df.values[:, 0:4] 
Y = iris.target

print(Y)
```

```
#printing samples and target 

X = df.values[:, 0:4] 
Y = iris.target

print(Y)
```

```
# Splitting the dataset into train and test 

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 100)
```

```
# Decision tree classifier 

clf= DecisionTreeClassifier(random_state = 100)

#fitting the training data

fig=clf.fit(X_train, y_train)
```

```
print('Decision Tree Classifer Created')
cn=['setosa', 'versicolor', 'virginica']
```

```
# prediction on random data

X=[[6.4,1.8 ,6.6 ,2.1]]

Y_pred=clf.predict(X)

print(Y_pred)
```

```
# prediction on X_test (testing data )

Y_pred=clf.predict(X_test)

print(Y_pred)
```

```
#Accuray of the model 

import sklearn.metrics as metrics
print("Accuracy:",metrics.accuracy_score(y_test, Y_pred))

cm=np.array(confusion_matrix(y_test,Y_pred))

cm
```

```
# classification metrics
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test, Y_pred))
```

```
# Import necessary libraries for graph viz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

dot_data = StringIO()
export_graphviz(clf, out_file=dot_data, feature_names=iris.feature_names,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
